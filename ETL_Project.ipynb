```python
import requests
import json
import pandas as pd
from sqlalchemy import create_engine
import requests
from pprint import pprint
from config import api_key
```

```python
medincome_csv = "Unemployment.csv"
med_income_df = pd.read_csv(medincome_csv, delimiter =",", encoding="utf-8-sig", skiprows=7)
med_income_df.head()
```

```python
medincome_cleaned = med_income_df.rename(columns={"Med_HH_Income_Percent_of_State_Total_2017":"Median Household Income 2013"})

medincome_cleaned
```

```python
fips = medincome_cleaned["FIPS"]
state_abb = medincome_cleaned["State"]
state_name = medincome_cleaned["Area_name"]
med_inc = medincome_cleaned["Median Household Income 2013"]
```

```python
medincome_complete1 = pd.DataFrame({
    "FIPS": fips,
    "State Abbreviation": state_abb,
    "State Name": state_name,
    "2013 Median Income": med_inc})
medincome_complete1.head(10)
```

```python
medincome_nj = medincome_complete1.loc[medincome_complete1['State Abbreviation'] == 'NJ']
medincome_nj
```

```python
medincome_nj2 = medincome_nj.iloc[1:]
medincome_nj2
```

```python
medincome_nj2['Currency']='USD' 
```
